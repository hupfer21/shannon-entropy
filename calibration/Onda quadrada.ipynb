```python
import numpy as np
import matplotlib.pyplot as plt
import math

def entropy(p):
    e=0
    b=200
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=p[i]*math.log(p[i],b)
    return e
def square(amp, freq, rc, ppc, c): #amplitude, frequencia, razao ciclica, pontos por ciclo, ciclos
    x=np.zeros(ppc*c)
    cont=0
    for i in range(ppc*c):
        if cont>=rc*ppc:
            x[i]=amp
        cont+=1
        if cont>ppc:
            cont=0
    return x
amp=1
freq=60
rc=0.9
ppc=100
c=2
cont=0

rc=0.1
e=np.zeros(9)
for cont in range(0,9):
    x=square(amp, freq, rc, ppc, c)
    t=np.linspace(0,c/freq,ppc*c)
    hist,bins=np.histogram(x,bins=2,density=0)

    p=np.zeros(ppc*c)
    for i in range(len(hist)):
        p[i]=hist[i]/(ppc*c)
    e[cont]=entropy(p)
    rc+=0.1
y=np.linspace(0,1,9)
plt.ylabel("Entropia")
plt.xlabel("Razão Cíclica")
plt.plot(y,e)

```
