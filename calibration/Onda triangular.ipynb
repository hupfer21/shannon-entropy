```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import math

def entropy(p):
    e=0
    b=100
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=p[i]*math.log(p[i],b)
    return e
f=60
rc=0.5
pontos=100
c=1

rc=0.01
e=np.zeros(99)
for cont in range(0,99):
    t=np.linspace(0,c/f,c*pontos)
    triang=signal.sawtooth(2*np.pi*f*t,rc)

    hist,bins=np.histogram(triang,bins=c*pontos,density=0)

    p=np.zeros(c*pontos)
    for i in range(len(hist)):
        p[i]=hist[i]/(c*pontos)
    e[cont]=entropy(p)
    rc+=0.01
y=np.linspace(0,1,99)
plt.ylabel("Entropia")
plt.xlabel("Razão Cíclica")
plt.plot(y,e)
```
