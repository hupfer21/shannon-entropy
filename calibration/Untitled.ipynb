```python
import numpy as np
import matplotlib.pyplot as plt
def probability(sin, amplitude): #paremetros: funcao senoidal, amplitude
    p=np.zeros(len(sin))
    for i in range(len(sin)):
        p[i]=1/(np.pi*np.sqrt(1-(sin[i]/amplitude)**2))
    return p
def entropy(p):
    e=0
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=p[i]*np.log2(p[i])
    return e
def test_entropy(max_cycles, samples_per_cycle, frequency, amplitude):
    entropies=np.zeros(int(max_cycles/1000))
    i=0
    for cycles in range(1, max_cycles, 1000):
        t=np.linspace(0, cycles/frequency, cycles*samples_per_cycle)
        sin=amplitude*np.sin(2*np.pi*frequency*t)
        p=probability(sin, amplitude)
        entropies[i]=entropy(p)
        i+=1   
    return entropies
c=1 #cycles
f=60 #frequency
spc=1000 #samples per cycle
amp=10 #amplitude
t=np.linspace(0,c/f,c*spc)
ruido=np.random.randn()
sin=amp*np.sin(2*np.pi*f*t)
p=probability(sin,amp)
x=np.linspace(-amp,amp,spc)
plt.plot(x,p)
plt.show()
entropy(p)
```
