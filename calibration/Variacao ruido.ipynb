```python
import numpy as np
import matplotlib.pyplot as plt
import math

def entropy(p):
    e=0
    b=1000
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=p[i]*math.log(p[i],b)
    return e

f=60 #frequencia
amp=5 #amplitude
pontos=1000

t=np.linspace(0,1/f,pontos)
seno=amp*np.sin(2*np.pi*f*t)

k=0
e=np.zeros(900)
for amp_r in np.arange(1,10,0.01):
    ruido=np.random.rand(pontos)
    sinal=amp_r*ruido
    hist,bin_edges=np.histogram(sinal,bins=pontos,density=0)

    p=np.zeros(pontos)
    for i in range(len(hist)):
        p[i]=hist[i]/pontos
    e[k]=entropy(p)
    k+=1
x=np.linspace(1,10,900)
plt.ylabel("Entropia")
plt.xlabel("Amplitude do Rudo")
plt.plot(x,e)
```
