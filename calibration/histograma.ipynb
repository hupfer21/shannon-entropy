```python
import numpy as np
import matplotlib.pyplot as plt
import math

def entropy(p):
    e=0
    b=1000
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=(p[i]*math.log(p[i],b))
    return e

f=60 #frequencia
amp=1 #amplitude
pontos=1000

t=np.linspace(0,1/f,pontos)
seno=amp*np.sin(2*np.pi*f*t)
hist,bin_edges=np.histogram(seno,bins=pontos,density=0)
 
p=np.zeros(pontos)
for i in range(len(hist)):
    p[i]=hist[i]/pontos
e=entropy(p)

ruido=np.random.randn(pontos)
sinal=seno+ruido
hist_ruido,bin_ed=np.histogram(sinal,bins=pontos,density=0)

pr=np.zeros(pontos)
for i in range(len(hist_ruido)):
    pr[i]=hist_ruido[i]/pontos
er=entropy(pr)

plt.plot(t,sinal,label='Sinal com Ruído, Entropia= {0:.6f}'.format(er))
plt.plot(t,seno,label='Sinal sem Ruído, Entropia= {0:.6f}'.format(e))
plt.legend()
```
