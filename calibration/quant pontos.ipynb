```python
import numpy as np
import matplotlib.pyplot as plt
import math

def entropy(p,b):
    e=0
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=p[i]*math.log(p[i],b)
    return e

f=60 #frequencia
amp=10 #amplitude
pontos=1000

e=np.zeros(100)
k=0
for pontos in range(1,100000,1000):
    t=np.linspace(0,1/f,pontos)
    seno=amp*np.sin(2*np.pi*f*t)
    hist,bin_edges=np.histogram(seno,bins=pontos,density=0)

    p=np.zeros(pontos)
    for i in range(len(hist)):
        p[i]=hist[i]/pontos
    e[k]=entropy(p,pontos)
    k+=1
x=np.linspace(1,10000,100)
plt.xscale("log")
plt.plot(x,e)
```
