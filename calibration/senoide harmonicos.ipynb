```python
import numpy as np
import matplotlib.pyplot as plt
import math

def entropy(p,b):
    e=0
    for i in range(len(p)):
        if p[i]!=0 and p[i]<1:
            e-=(p[i]*math.log(p[i],b))
    return e

pontos=1000
t=np.linspace(0,2,pontos)
har=np.zeros(pontos)
freq=2
cont=0
e=np.zeros(10)
for j in range(10):
    for f in range(1,freq):
        har+=np.sin(2*np.pi*f*t)

    hist,bin_edges=np.histogram(har,bins=pontos,density=0)

    p=np.zeros(pontos)
    for i in range(len(hist)):
        p[i]=hist[i]/pontos
    e[cont]=entropy(p,pontos)
    cont+=1
    freq+=1
x=np.linspace(1,10,10)
plt.xlabel("nÂ° de harmonicos com freq +1")
plt.ylabel("entropia")
plt.plot(x,e)
```
